Sigmoid Function:

f(x)
= 1/(1+e^-x)

```python

def sigmoid_function(x): 
    """
    This function takes as input a floating point number and outputs the sigmoidal value of that input. 
    """
    return 1/(1+np.exp(-x))

def approximated_sigmoid_function( x ):
    """
    This function takes as input a floating point number and outputs the approx sigmoidal value of that input. 
    """
    return ( (1.0/2) + (1.0/4)*x - (1.0/48)*x**3 + (1.0/480)*x**5)

def sigmoid_der( x ): 
    return sigmoid_function(x)*(1-sigmoid_function(x))

def approximated_sigmoid_deriavative( x ):
    return approximated_sigmoid_function(x)*(1-approximated_sigmoid_function(x))
```

```python
x = [ ]
original_sigmoids = []
approximated_sigmoids = []
for ii in np.linspace(-5, 5 , 200 ):
    x.append(ii)
    original_sigmoids.append(sigmoid_function(ii))
    approximated_sigmoids.append(approximated_sigmoid_function(ii))
```

```python
for ii in range(0, 200): 
    print x[ii] ,'\t','\t', original_sigmoids[ii] ,'\t', approximated_sigmoids[ii] 
```

```python
plt.plot(x,approximated_sigmoids )
plt.plot(x,original_sigmoids )
```
